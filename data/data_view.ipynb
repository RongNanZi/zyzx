```python
my_text = []
with open('filtered_sms_sample.txt') as f:
    for i in range(200):
        my_text.append(f.readline())
```

```python
my_text[:3]
```

```python
ner_dict = {'sence':u'【(?P<sence>.+)】',
            'time':u'(?P<time>(\d{4}年)?\d{1,2}月\d{1,2}日\d{1,2}[时点]\d{1,2}分(\d秒)?)', 
            'consume':u'消费为?是?：?(?P<consume>.+?[元角分￥])',
            'balance':u'余额[为是]：?(?P<balance>.+?[元分角￥])',
            'used_tranffic':u'已使用.*(流量)+(?P<used_tranffic>\d+.?\d+?[MG])'}
import re

new_line = my_text[0]
for k, re_v in ner_dict.items():
    find = re.search(re_v, my_text[0])
    if find: 
        print(find.group(k))
        new_line = re.sub(re_v, u'${%s:\g<%s>}'%(k,k), new_line)
    else:
        print('not find')
```

```python
import pandas as pd
d = pd.DataFrame(data = {'key' : list(ner_dict.keys()),
                         'cn_key':[u'场景',u'时间',u'消费金额',u'余额',u'使用流量'],
                         're':list(ner_dict.values())})
```

```python
d.to_csv('sqd_kre.csv', index=False)
```
