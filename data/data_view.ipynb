```python
my_text = []
with open('filtered_sms_sample.txt') as f:
    for i in range(200):
        my_text.append(f.readline())
```

```python
my_text[:3]
```

```python
ner_dict = {'sence':u'【(?P<sence>.+)】',
            'time':u'(?P<time>(\d{4}年)?\d{1,2}月\d{1,2}日\d{1,2}[时点]\d{1,2}分(\d秒)?)', 
            'consume':u'消费为?是?：?(?P<consume>.+?[元角分￥])',
            'balance':u'余额[为是]：?(?P<balance>.+?[元分角￥])',
            'used_tranffic':u'已使用.*(流量)+(?P<used_tranffic>\d+.?\d+?[MG])'}
import re

new_line = my_text[0]
for k, re_v in ner_dict.items():
    find = re.search(re_v, my_text[0])
    if find: 
        print(find.group(k))
        new_line = re.sub(re_v, u'${%s:\g<%s>}'%(k,k), new_line)
    else:
        print('not find')
```

```python
import re
all_rank = pd.read_table('all.rank', sep=' ',header=None, names=['id', 'key', '频度'])
#如果key全为数字，则标为弃用, 频度为1的Key弃用    
def filter(x):
    if x['频度'] == 1 or re.search('^\d+\.?$', x['key']):
        return 1
    return 0
all_rank['弃用'] = all_rank.apply(filter, axis=1)
filtered_all_rank = all_rank.drop([i for i in all_rank.index if all_rank.iloc[i]['弃用']==0])
filtered_all_rank = filtered_all_rank.drop(['弃用'], axis=1)
filtered_all_rank.to_csv('filter_all_rank.csv', index=False)
```
