```python
import pandas as pd
check_data = pd.read_excel('北京中移在线/算法识别结果校验情况.xlsx')
```

```python
check_data
```

```python
check_keys =[]
def get_key(x):
    global check_keys
    x = x.strip('（').split('）（')
    for item in x:
        item = item.split('：')
        if item[0] not in check_keys:
            check_keys.append(item[0])
        
check_data['待提取重要参数'].dropna().apply(get_key)
```

```python
'订单状态' in check_keys
```

```python
all_keys = pd.read_excel('北京中移在线/原生模版配置关键字V1.1.xlsx')
shoud_key = all_keys['短信模板中对应配置'].drop_duplicates().apply(lambda x: x[4:]).values
```

```python
outline_keys = []
for i in check_keys:
    if i not in shoud_key:
        outline_keys.append(i)
```
