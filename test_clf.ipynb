```python
import pandas as pd
from clf import *

c_rule = { "10049积分查询":{"当前积分":1,"新增积分":1,"剩余积分":1, "兑换":1,\
                        "失效时间":1,"到期时间":1,"积分":1,"【积分兑换提醒】":2, "可兑换积分":2},\
          "10052充值成功":{"缴费金额":1,"交费时间":1,"缴费时间":1,"缴费渠道":1,"交费渠道":1,\
                    "充值":1,"交易订单号":1,"交易流水号":1},\
          "10055账单提醒":{"消费":1,"账单":1,"账户余额":1, "备注":1}}
data = pd.read_excel('data/10086sms_log.xlsx', sheet_name='Sheet1')
for sence_name,_ in c_rule.items():
    data['{}_score'.format(sence_name)] = [0]*data.shape[0]
```

```python
def get_score(df, c_rule):
    for sence_name, s_rule  in c_rule.items():          
            sence_score = 0                             
            for token, t_score in s_rule.items():       
                    if df[0].find(token) != -1:      
                            sence_score += t_score      
            df['{}_score'.format(sence_name)] = sence_score
    return df
data = data.apply(get_score, args=([c_rule]), axis=1)
```

```python
sence_threshold = {"10049积分查询_score":1, "10052充值成功_score":1, "10055账单提醒_score":1}

def get_class(s, sence_threshold):
    
    wanted = s[:1]
    #get the max classification score
    c_s = s[1:]
    max_sence = c_s.index[c_s.values.argmax()]
    if c_s[max_sence] > sence_threshold[max_sence]:
        wanted['sence'] = max_sence
    else:
        wanted['sence'] = None
    return wanted
            

after_classification = data.apply(get_class, args=([sence_threshold]), axis=1)
```

```python
after_classification = after_classification.dropna()
def get_filter(df):
    s = df.iloc[0][0]
    
    new_s = ''.join([i for i in s if not i.isdigit() | (i == '-')])
    #ser['短句'] = new_s.split('.。；;!！')
    wanted = pd.DataFrame()
    wanted['短句'] = new_s.split('.。；;!！')
    wanted['原文ID'] = [df.index[0]]*wanted.shape[0]
    wanted['原句'] = [s]*wanted.shape[0]
    wanted['sence'] = [df.iloc[0][1]]*wanted.shape[0]
    
    return wanted
        
after_filed = after_classification.groupby('原文').apply(get_filter)
```

```python
after_filed.drop_duplicates(['短句']).to_csv('after_filter.csv', index=False)
```
