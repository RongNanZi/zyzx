```python
import pandas as pd
csv_data = pd.read_excel('data/sqd_manual_confirm.xlsx', sheet_name="话费类")
```

```python
cn_key_set = []
with open('data/lidongdong/template_Phone_key.txt') as f:
    for line in f.readlines():
        cn_key_set.append(line.strip())
def get_we_find(s):
    wanted = {}
    items = s.split(',')
    for item in items:
        kv = item.split(':')
        wanted.update({kv[0].strip('{\'\' '):kv[1].strip('}\' ')})
    return wanted
def get_they_find(s):
    wanted = {}
    print(s)
    items = s.split('）（')
    for item in items:
        kv = item.split('：')
        wanted.update({kv[0].strip(' （'): kv[1].strip(' ）')})
    return wanted
```

```python
for i in csv_data.index[51:64]:
    repeat_item , not_in= [], []
    advice = ''
    we_find = csv_data.loc[i]['we_find']
    we_find = get_we_find(we_find)
    they_find = csv_data.loc[i]['补充参数']
    if str(they_find) == 'nan':
        continue
    they_find = get_they_find(they_find)
    for item  in they_find.items():
        if item[0] in we_find.keys() and we_find[item[0]] == item[1]:
            repeat_item.append(item[0])
        elif item[0] not in cn_key_set:
            not_in.append(item[0])
    
    if len(repeat_item)!=0:
        advice += '\"{}\"已包含；'.format(repeat_item)
    if len(not_in)!=0:
        advice +=  '需求文档中未出现\"{}\"；'.format(not_in)
    csv_data.loc[i]['开发建议'] = advice 
```
